```python
import tkinter as tk

# Function to evaluate the expression
def evaluate_expression(event=None):
    try:
        result = str(eval(entry.get()))
        entry.delete(0, tk.END)
        entry.insert(tk.END, result)
    except:
        entry.delete(0, tk.END)
        entry.insert(tk.END, "Error")

# Function to handle button click
def on_button_click(value):
    entry.insert(tk.END, value)

# Function to clear the display
def clear_display():
    entry.delete(0, tk.END)

# Function to handle backspace
def backspace():
    current = entry.get()
    entry.delete(0, tk.END)
    entry.insert(tk.END, current[:-1])

# Main window
root = tk.Tk()
root.title("Python Calculator")
root.geometry("500x400")
root.resizable(False, False)

entry = tk.Entry(root, font=("Arial", 30), borderwidth=8, relief="ridge", justify="left")
entry.pack(fill="both", ipadx=8, ipady=15, padx=10, pady=10)

# Buttons
buttons = [
    ['7', '8', '9', '/'],
    ['4', '5', '6', '*'],
    ['1', '2', '3', '-'],
    ['0', '.', '=', '+'],
]

frame = tk.Frame(root)
frame.pack()

for row in buttons:
    r = tk.Frame(frame)
    r.pack(expand=True, fill="both")
    for btn in row:
        action = evaluate_expression if btn == '=' else lambda x=btn: on_button_click(x)
        tk.Button(r, text=btn, font=("Arial", 20), command=action).pack(side="left", expand=True, fill="both")

# Clear and backspace buttons
control_frame = tk.Frame(root)
control_frame.pack(fill="both", expand=True, padx=10, pady=(5, 10))

tk.Button(control_frame, text="C", font=("Arial", 16), bg="tomato", fg="white", command=clear_display).pack(side="left", expand=True, fill="both", padx=10)
tk.Button(control_frame, text="âŒ«", font=("Arial", 16), bg="gray", fg="white", command=backspace).pack(side="left", expand=True, fill="both", padx=10)

# Keyboard bindings
root.bind("<Return>", evaluate_expression)
root.bind("<BackSpace>", lambda e: backspace())
root.bind("<Escape>", lambda e: clear_display())

# Allow typing numbers and operators
def handle_keypress(e):
    if e.char in '0123456789+-*/.':
        entry.insert(tk.END, e.char)

root.bind("<Key>", handle_keypress)

# Run app
root.mainloop()
```
