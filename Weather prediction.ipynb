```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the weather dataset
df = pd.read_csv("Date,Temperature,Humidity,WindSpeed 2.csv")

# Display basic info
print("ðŸ“Š Weather Dataset Summary:")
print(df.describe())
print("\nðŸŒ¦ Weather Counts:")
print(df['Weather'].value_counts())

# Plot temperature over time
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Humidity'], marker='o', color='Purple')
plt.title("Daily Temperature")
plt.xlabel("Date")
plt.ylabel("Temperature (Â°C)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Bar chart of weather type counts
df['Weather'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Weather Type Frequency")
plt.xlabel("Weather")
plt.ylabel("Count")
plt.grid(axis='y')
plt.tight_layout()
plt.show()


```
